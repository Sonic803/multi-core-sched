```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

def exponential_pdf(x, lambd):
    return lambd * np.exp(-lambd * x) * (x >= 0)  # PDF of an exponential distribution

# Define parameters
lambda1 = 0.2  # Mean = 1/lambda
lambda2 = 0.5  # Mean = 1/lambda

# Define range
x = np.linspace(0, 10, 1000)
exp1 = exponential_pdf(x, lambda1)
exp2 = exponential_pdf(x, lambda2)

# Compute convolution numerically
dx = x[1] - x[0]  # Step size
conv_result = convolve(exp1, exp2, mode='full') * dx

# Compute new x-axis for convolution
x_conv = np.linspace(0, 2 * x[-1], len(conv_result))

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(x, exp1, label=f'Exponential PDF (λ={lambda1})', linestyle='--')
plt.plot(x, exp2, label=f'Exponential PDF (λ={lambda2})', linestyle='--')
plt.plot(x_conv, conv_result, label='Convolution', color='red')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Convolution of Two Exponentials')
plt.legend()
plt.grid()
plt.show()
```
